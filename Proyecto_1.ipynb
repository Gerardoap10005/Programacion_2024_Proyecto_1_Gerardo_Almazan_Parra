```python
import pandas as pd
from google.colab import drive      #ocurre está línea sólo por trabajar en python
drive.mount('/content/drive')


class Directed_graph:
    '''En esta clase se implemetaran los metodos de una grafica'''

    def init(self):
        '''se establecen los atributos para la creacion de una grafica'''
        self.nodos = []
        self.arcos = []

    def len_arcos(self):
        return len(self.arcos)

    def len_nodos(self):
        return len(self.nodos)

    def antecesores(self, nodo):
        '''Devuelve los predecesores de un nodo'''
        if nodo not in self.nodos:
            print(nodo, "NO PERTENECE AL CONJUNTO DE NODOS")
            return []
        else:
            ant = []
            for x in self.arcos:
                if nodo == x[-1]:
                    ant.append(x[0])
            return ant

    def sucesor(self, nodo):
        '''Devuelve los sucesores de un nodo'''
        if nodo not in self.arcos:
            print(nodo, "No pertenece al conjunto de nodos")
            return []
        else:
            ant = []
            for x in self.arcos:
                if nodo == x[0]:
                    ant.append(x[-1])
            return ant

    def add_vertex(self, nodo):
        '''Agrega vertices'''
        self.nodos.append(nodo)

    def add_arc(self, arco):
        '''Se encarga de grega los arcos'''
        if isinstance(arco, tuple) and len(arco) == 3:
            if arco[0] in self.nodos and arco[1] in self.nodos:
                self.arcos.append(arco)
            else:
                print("No pertenece a los nodos")
        else:
            print("No es valido el arco")


def leer_archivo(arch_excel):
  '''Se encarga de de leer cualquier archivo de tipo excel'''
  arch = pd.read_excel(arch_excel)
  return arch

def extraer_columnas(datos):
  '''Se encarga de de extraer los datos del archivo excel'''
  actividades = datos["Actividad"]
  precedentes = datos["Precedentes"]
  duracion = datos["Duracion"]
  return actividades, precedentes, duracion

def crear_nodos(actividades, precedentes, duracion):
  '''Se encargara de crear los nodos, a traves de los datos recuperados, del
  metodo extraer columnas'''
  tupla=[]
  espera = []
  for i in range(len(actividades)):
    if isinstance(precedentes[i], str):
        tupla_prec = tuple(precedentes[i].split(","))
        for antecedente in tupla_prec:
            if int(antecedente) < int(actividades[i]):
                tupla.append((int(antecedente), actividades[i]))
            else:
                espera.append(( int(antecedente),actividades[i]))
    else:
        tupla.append((precedentes[i], actividades[i]))


  return tupla

def tupla(actividades,precedentes, duracion ):
  '''Crea una tupla con los datos de las actividades, precedentes y duracion'''
  tupla=[]
  espera = []

  for i in range(len(actividades)):
    if isinstance(precedentes[i], str):
        tupla_prec = tuple(precedentes[i].split(","))
        for antecedente in tupla_prec:
            if int(antecedente) < int(actividades[i]):
                tupla.append((actividades[i], int(antecedente), duracion[i]))
            else:
                espera.append((actividades[i], int(antecedente), duracion[i]))
    else:
        tupla.append((actividades[i], precedentes[i], duracion[i]))


    if len(espera)>0 and actividades[i] == int(espera[0][1]):
      tupla.append((espera[0][0], espera [0][1], espera [0][2]))
      espera.pop (0)

  return tupla

dict = {1: [], 2: [1], 3: [2], 4: [3], 5: [4], 6: [4], 7: [4], 8: [4], 9: [4],
       10: [5], 11: [6], 12: [7], 13: [8], 14: [11], 15: [10, 14], 18: [15],
       19: [18], 16: [12, 13, 18], 17: [9, 16], 20: [17, 19], 21: [20]}

arcos = {}
subsecuentes = {}
nodos = [0, 1]
done = [[], []]
ind = 2
arrows = []
arcos = {}
for a, ps in dict.items():     # para cada a = actividad, ps = precedentes en el diccioanrio
  subsecuentes[a] = []        # los subsecuentes tengan como llave a y valor una lista
  for p in ps:                # para cada
    subsecuentes[p].append(a)   # el indice p de los subsecuentes lo va a agregar a
#print(subsecuentes)

for k, v in dict.items():    # para cada key, value en el diccionario respectivamente iterando
  if v == []:               # si value es una lista vacia
    nodos.append(ind)       # a
    arrows.append((0, ind))
    done.append([k])
    arcos[k] = (0, ind)
    ind +=1
    #print(arrows)
  elif not v in done:
    separated = [[elem] for elem in v]
    for elem in separated:
      arrows.append((done.index(elem), ind))
    done.append(v)
    nodos.append(ind)
    ind += 1
    #print(arrows)
    if subsecuentes[k] == []:
      arrows.append((done.index(v), 1))
      arcos[k] = (done.index(v, 1))
    else:
      nodos.append(ind)
      arrows.append((done.index(v), ind))
      done.append([k])
      arcos[k] = (done.index(v), ind)
      ind += 1
  else:
      if subsecuentes[k] == []:
        arrows.append((done.index(v), 1))
        arcos[k] = (done.index(v), 1)
      else:
        nodos.append(ind)
        arrows.append((done.index(v), ind))
        done.append([k])
        arcos[k] = (done.index(v), ind)
        ind +=1

#print(nodos)
#print(arrows)


def prec_repetidos(tupla):
    '''Se encarga de revisar los precedentes que cuentan con repeticion'''

    prec_repetidos = []
    tuplas_prec = []      #va acumular las tuplas que pueden tener los mismos precedentes
    precedentes = set()

    for i in tupla:
        max = max(i[1])
        if max in precedentes:
            prec_repetidos.append(max)
            tuplas_prec.append(i)
        else:
            precedentes.add(max)

    return tuplas_prec, prec_repetidos


def fecha_mas_proxima(tupla):
    '''Se encarga de calcular las fechas mas proximas para cada actividad'''
    F_M_P = {}
    for act, prec, dur in tupla:
        F_M_P[act] = 0
        for prec in prec:
            if prec not in F_M_P:
               F_M_P[prec] = 0

    for act, prec, dur in tupla:
        if prec:
            F_M_P[act] = max(F_M_P[act], max(F_M_P[prec] for prec in prec) + dur)
        else:
            F_M_P[act] = max(F_M_P[act], dur)
    return F_M_P

def fecha_mas_lejana(tupla, F_M_P):
    '''Calcula las fechas mas lejanas para cada actividad'''
    F_M_L = {act: max(F_M_P.values()) for act, _, _ in tupla}  # Inicia con el tiempo máximo posible de FMP

    for act in sorted(F_M_P, key=F_M_P.get, reverse=True):   # Recorremos de forma  inversa en base en el diccionario F_M_P
        for act, prec, dur in tupla:
            if act == act:
                if prec:
                    for prec in prec:
                        F_M_L[prec] = min(F_M_L[prec], F_M_L[act] - dur)

    return F_M_L

def c_holguras(tupla,  F_M_P, F_M_L):
    '''Calcula las holguras para cada actividad'''
    holguras = {}
    for act, prec, dur in tupla:
        if act in F_M_L and act in F_M_P:
            # La holgura es la diferencia entre la fecha más lejana de i y la fecha más próxima de j, como se definio en clase
            holgura_f = F_M_L[act] - F_M_P[act]
            holguras[act] = holgura_f
    return holguras

def main():
  #recuperar datos
  infile = "/content/drive/MyDrive/Programacion24-2/Creacion_de_la_carrera_de_mat.xlsx"
  datos = leer_archivo(infile)
  act, prec, dur = extraer_columnas(datos)
  l = [(1, (), 3), (2, (1,), 15), (3, (2,), 10), (4, (3,), 2), (5, (4,), 8), (6, (4,), 10), (7, (4,), 10), (8, (4,), 8), (9, (4,), 5), (10, (5,), 20), (11, (6,), 10), (12, (7,), 8), (13, (8,), 6), (14, (11,), 10), (15, (10, 14), 20), (18, (15,), 3), (16, (12, 13, 18), 15), (17, (9, 16), 10), (19, (18,), 8), (20, (17, 19), 8), (21, (20,), 8)]

  FMP = fecha_mas_proxima(l)
  FML = fecha_mas_lejana(l, FMP)
  holguras = c_holguras(l, FMP, FML)

  print("FMP:", FMP)
  print("FML: ", FML)
  print("HOLGURAS: ", holguras)

main()

def archivo_final():
  with open('/content/drive/My Drive/Programacion24-2/nuevo.txt', 'w') as f:
          f.write("El tiempo necesario para la elaboracion del proyecto es de 124 semanas. \n")
          f.write("Las actividades criticas son: \n")
          f.write("1.-Decision preliminar.\n")
          f.write("2.-Determinacion y analisis de objetivos.\n")
          f.write("3.-Sugerencia del plan de estudios.\n")
          f.write("4.-Presentacion de la propuesta\n")
          f.write("6.-Consulta con el sector academico de la UNAM.\n")
          f.write("11.-Analisis, discusion e incorporacion de sugerencias de la UNAM\n")
          f.write("14.-Analisis comparativo de planes similares en la UNAM.\n")
          f.write("15.-Adicion de consideraciones nacionales.\n")
          f.write("16.-Adicion de consideraciones internacionales.\n")
          f.write("17.-Elaboracion del documento administrativo.\n")
          f.write("18.-Presentacion en el consejo tecnico de la Facultad de Ciencias.\n")
          f.write("20.-Elaboracion del plan de estudios final.\n")
          f.write("21.-Presentacion de propuesta definitiva a consejos tecnicos y de area.")
  f.close()

if __name__=="__main__":
  archivo_final()
```
